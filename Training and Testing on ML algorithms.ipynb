# Train Dataset

```python
import pandas
Ham_Spam_All=pandas.read_csv("train_trec07p_Ham_Spam_All.csv")
```

```python
print(Ham_Spam_All.shape)
```

```python
Ham_Spam_All.head()
```

# By getting Text features and Class

```python
Text_Features_train=Ham_Spam_All.Ham_Spam_email_text
Class_train=Ham_Spam_All.Class
```

# Features Extraction by using TFIDF

Features extraction from ham and spam emails for training the machine learning model
- tokenizing
- ngram
- removing stop words 


# - Unigram features

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    stop_words='english',
    ngram_range=(1, 1),
    max_features =1000)

```

# Preparing features for training and testing

by using word vectorisor to fit transform on all words of emails 

```python
unigramdataGet= word_vectorizer.fit_transform(Text_Features_train)
unigramdataGet = unigramdataGet.toarray()
vocab = word_vectorizer.get_feature_names()
unigramdata=pd.DataFrame(np.round(unigramdataGet, 1), columns=vocab)
unigramdata[unigramdata>0] = 1

unigramdata.head(20)
```

# - Bigram features

# - Trigram features

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    stop_words='english',
    ngram_range=(3, 3),
    max_features =1000)
```

by using word vectorisor to fit transform on all words of emails 

```python
TrigramdataGet= word_vectorizer.fit_transform(Text_Features_train)
TrigramdataGet= TrigramdataGet.toarray()

vocab1 = word_vectorizer.get_feature_names()
Trigramdata=pd.DataFrame(np.round(TrigramdataGet, 1), columns=vocab1)
Trigramdata[Trigramdata>0] = 1
Trigramdata.head(20)
```

# Machine learning training for Unigram

# SVM Algorithm

```python
from sklearn.svm import LinearSVC


svc_uni=LinearSVC(random_state=0, tol=1e-5)
svc_uni = svc_uni.fit(unigramdata , Class_train)
svc_uni
```

# Logistic regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
Lo_Reg_uni= LogisticRegression(penalty = 'l2', C = 1)
Lo_Reg_uni = Lo_Reg_uni.fit(unigramdata , Class_train)
Lo_Reg_uni
```

# RandomForestClassifier Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
Ran_For_uni = RandomForestClassifier(n_estimators=100,max_depth=15, random_state=42,max_leaf_nodes=10)
Ran_For_uni = Ran_For_uni.fit(unigramdata , Class_train)
Ran_For_uni
```

# BernoulliNB Algorithm

```python
from sklearn.naive_bayes import BernoulliNB
Br_NB_uni = BernoulliNB(alpha=1.0, binarize=0.0)
Br_NB_uni = Br_NB_uni.fit(unigramdata , Class_train)
Br_NB_uni
```

# Now training all data (training&testing) on BernoulliNB  Algorithm

```python
import pandas
Ham_Spam_All1=pandas.read_csv("train_trec07p_Ham_Spam_All.csv")
Ham_Spam_All2=pandas.read_csv("test_trec07p_Ham_Spam_All.csv")

Ham_Spam_All3=pandas.concat([Ham_Spam_All1,Ham_Spam_All2])
Ham_Spam_All3.to_csv("All_Trec_Ham_Spam_All.csv", index=False)
Ham_Spam_All3.head(4)
```

```python
X=Ham_Spam_All3.Ham_Spam_email_text
Y=Ham_Spam_All3.Class
```

# - Trigram features

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    stop_words='english',
    ngram_range=(3, 3),
    max_features =1000)


TrigramdataGet= word_vectorizer.fit_transform(X)
TrigramdataGet= TrigramdataGet.toarray()

vocab1 = word_vectorizer.get_feature_names()
Trigramdata=pd.DataFrame(np.round(TrigramdataGet, 1), columns=vocab1)
Trigramdata[Trigramdata>0] = 1
Trigramdata.head(20)
```

```python
from sklearn.naive_bayes import BernoulliNB
Br_NB_tri = BernoulliNB(alpha=1.0, binarize=0.0)
Br_NB_tri = Br_NB_tri.fit(Trigramdata , Y)
Br_NB_tri
```

# By Entring the Ham email text

```python
he = input("Please enter email text here: ")
newF=pd.DataFrame(
    {
        "He":[he]
    }
)
```

```python
x = word_vectorizer.transform(newF.He)
```

```python
aa=Br_NB_tri.predict(x)
if aa==1:
    print("Ham")
else:
    print("Spam")
```
